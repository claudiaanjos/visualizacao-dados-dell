```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt'
df = pd.read_csv(uri)
df.head()
```

2. Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK?

```python
plt.figure(figsize=(20,15))
plt.title('Variação das vendas ao longo dos meses')
sns.lineplot(x="Month", y="Sales", data=df)
plt.xlabel('Período (ano-mês)', fontsize=20)
plt.ylabel('Vendas', fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

Interessante observar que as vendas aumentaram ao longo desses três anos.
