````python
#importando os dados
df=pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv')
````

````python
#Conhecendo o dataset
df.head(5)
````

1. Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?

````python
#Média
df.mean()
````

````python
#Desvio padrão
df.std()
````

2. Qual o consumo de licor na França?

````python
df_consumo_franca = df[df['country'] == 'France']
````

````python
#Resposta
df_consumo_franca['spirit_servings'].values[0]
````

3. Quais os 5 países nos quais se consome mais vinho?

````python
#Resposta
df_ordenado_vinho = df.sort_values(by='wine_servings', ascending = False).head(5)
df_ordenado_vinho
````

4. Quais os países nos quais o consumo de cerveja está acima da média?

````python
#Resposta
df[df['beer_servings'] > df['beer_servings'].mean()]
````

5. Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?

````python
#Verificando o consumo total de álcool do Brasil
consumo_brasil = df[df['country'] == 'Brazil']
consumo_brasil_total = consumo_brasil['total_litres_of_pure_alcohol'].values[0]
consumo_brasil_total 
````

````python
#Resposta
df[df['total_litres_of_pure_alcohol'] > consumo_brasil_total]
````
