1. Importe as bibliotecas pandas, matplotlib e seaborn, usando a palavra mágica inline, para exibir as visualizações no notebook. Carregue os dados, que podem ser carregados pelo seaborn, utilizando a função load_dataset(‘mpg’), em um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = sns.load_dataset("mpg")
df.head()
```

2. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!

Vamos responder por partes.

*Existe uma correlação entre a aceleração e o consumo? De que tipo?*

Resposta: Sim, conforme tabela apresentada no curso, entende-se que existe uma baixa correlação positiva.

***Tabela de correlações apresentada no curso e guia para responder as questões.***

image.png

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível dos veículos')
sns.regplot(x="acceleration", y="mpg", data=df)
plt.show()
```

*A correlação é a mesma para os diferentes países de origem?*

Resposta: Sim, conforme tabela apresentada no curso, entende-se que existe uma baixa correlação positiva nos EUA e Japão, mas não na Europa.

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível dos veículos pelo país de origem')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=df)
plt.show()
```

```python
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=df)
```

3. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

Existe alguma relação entre as três variáveis?

Resposta: Sim, uma baixa correlação positiva. Aparentemente, quanto mais novo o modelo do carro maior o consumo de combustível e aceleração.

```python
plt.figure(figsize=(10,7))
plt.title('Relação entre aceleração e consumo de combustível dos veículos pelo ano do modelo')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=df)
plt.show()
```
