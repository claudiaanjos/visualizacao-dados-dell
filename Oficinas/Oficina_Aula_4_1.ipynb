```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head(2)
```

2. Crie uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Você pode escolher a técnica que julgar mais adequada;

```python
plt.figure(figsize=(15,10))
plt.title('Valor da passagem de acordo com a cidade de embarque')
sns.stripplot(x="embark_town", y="fare", data=titanic)
plt.xlabel("Cidade de embarque")
plt.ylabel("Valor da passagem")
plt.show()
```
